2. Weighted Moving Average
3. Exponential Moving Average
4. Exponential Smoothing

```python
# Importing libraries
from __future__ import absolute_import,division,print_function,unicode_literals

import pandas as pd
import numpy as np
from datetime import datetime

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = (8,6)
mpl.rcParams['axes.grid'] = False
```

```python
# Importing electricity consumption data
ec_df = pd.read_csv('data/electricity_consumption.csv', parse_dates=['Bill_Date'])
ec_df.head()
```

```python
# Setting time column as index for easier time series analysis
ec_indexed = ec_df.set_index('Bill_Date')
ec_indexed.head()
```

```python
# Billed amount for 3 years
ec_2018 = ec_indexed['2016':'2018'][['Billed_amount']]
ec_2018
```

### Simple Moving Average 

The Simple Moving Average (SMA) is calculated by adding the price of an instrument over a number of time periods and then dividing the sum by the number of time periods. The SMA is basically the average price of the given time period, with equal weighting given to the price of each period.

In financial applications, a simple moving average (SMA) is the unweighted mean of the previous n data. However, in science and engineering, the mean is normally taken from an equal number of data on either side of a central value. This ensures that variations in the mean are aligned with the variations in the data rather than being shifted in time.

Mathmatically (t+(t-1)+(t-2)+...+(t-n))/n

Moving Average cant be a great tool when, the data is not stationary and fluctuating.

```python
# Using rolling mean of past 3 data points
ec_2018['ma_rolling_3'] = ec_2018['Billed_amount'].rolling(window=3).mean().shift(1)
```

```python
ec_2018
```

```python
ec_2018.plot();
```

### Weighted Moving Average 

Weighted moving averages can find trens sooner than SMA, on the other hand its complex as we need to assign the weights manually.

Weighted Moving Average = ((t\*weighing factor)+((t-1)\*weighing factor-1)+...+((t-n)\*weighing factor-n)))/n

```python
def wma(weights):
    def calc(x):
        return (weights*x).mean()
    return calc
```

```python
# Using weighted rolling mean of past 3 data points
ec_2018['wma_rolling_3'] = ec_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5]))).shift(1)
```

```python
ec_2018
```

```python
ec_2018.plot();
```

### Exponential Moving Average

It adapts quickly to the data point changes and we don't have to decide the weights manually.

Exponential Moving Average = (Close - Previous EMA) \* (2/(span+1)) + Previous EMA

```python
# Using exponential moving average of past 3 data points
ec_2018['ewm_window_3'] = ec_2018['Billed_amount'].ewm(span=3, adjust=False, min_periods=0).mean().shift(1)
```

```python
ec_2018
```

```python
ec_2018.plot();
```

### Exponential Smoothing

It Requires a parameter called alpha aka Smoothing parameter

Larger value of alpha means the model is paying attention to the newer values,smaller value of alpha mean model is giving importance to history value.

```python
# Using exponential smoothing of past 3 data points with alpha=0.7
ec_2018['esm_window_3_7'] = ec_2018['Billed_amount'].ewm(alpha=0.7, adjust=False, min_periods=3).mean().shift(1)
```

```python
ec_2018
```

```python
ec_2018[['Billed_amount', 'esm_window_3_7']].plot();
```

```python
# Using exponential smoothing of past 3 data points with alpha=0.3
ec_2018['esm_window_3_3'] = ec_2018['Billed_amount'].ewm(alpha=0.3, adjust=False, min_periods=3).mean().shift(1)
```

```python
ec_2018[['Billed_amount', 'esm_window_3_3']].plot();
```

```python
ec_2018.plot();
```

### Comparison

Lets compare RMSE values of all types of the moving averages

```python
# Simple Moving Average
((ec_2018['Billed_amount']-ec_2018['ma_rolling_3'])**2).mean()**0.5
```

```python
# Weighted Moving Average
((ec_2018['Billed_amount']-ec_2018['wma_rolling_3'])**2).mean()**0.5
```

```python
# Exponential Moving Average
((ec_2018['Billed_amount']-ec_2018['ewm_window_3'])**2).mean()**0.5
```

```python
# Exponential Smoothing with alpha=0.7
((ec_2018['Billed_amount']-ec_2018['esm_window_3_7'])**2).mean()**0.5
```

```python
# Exponential Smoothing with alpha=0.3
((ec_2018['Billed_amount']-ec_2018['esm_window_3_3'])**2).mean()**0.5
```

RMSE for Exponential Smoothing performed the best for our highly fluctuating dataset.
